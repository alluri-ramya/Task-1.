```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.strftime('%d-%m-%Y')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.strftime('%d-%m-%Y')
df
#Date format is changed 
```

```python
t = df.select_dtypes(include='object')
p=t.head()

#Text data columns have been extracted and named as 'p'
```

```python
from sklearn.preprocessing import LabelEncoder
d=pd.get_dummies(p)
d

#Categorial data is converrted to labelled data
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns

#the column names have been changed to lower case and the spaces have been replaced with '_' 
```

```python
int_cols = ['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']
df[int_cols] = df[int_cols].astype('Int64')

#the datatype of the above columns is set to integer
```
